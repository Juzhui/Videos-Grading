```python
#

#
from os import listdir,path

rootdir="raw"

name_list=[]
path_list=[]
factors_list=[]
    
file_list = listdir(rootdir) #

for i in range(0,len(file_list)):
    tmp_path = path.join(rootdir,file_list[i])
    (tmp_name, tmp_extension) = path.splitext(file_list[i])
    tmp_factors=tmp_name.split('_')
    if path.isfile(tmp_path):
        #print(tmp_path)
        path_list.append(tmp_path)
        name_list.append(tmp_name)
        factors_list.append(tmp_factors)

n=len(path_list)
print(n)

print(path_list)
print(name_list)
print(factors_list)
```

```python
#
import jieba
import re
import csv

#
stopwords = [line.strip() for line in open('stopwords.txt').readlines()]

#
outdir='wordcut'

for i in range(n):
    
    if factors_list[i][2]!='head':
        
        deal_raw=""
    
        filepath=path_list[i]
        f_in = open(filepath,'r',encoding='utf-8') #osopen
        raw=f_in.readlines()
        #print(raw)
        
        #commentdanmu
        for line in raw:
            try:
                tmp_sentence=line.split(',')
                deal_raw=deal_raw+tmp_sentence[2]
            except:
                print(name_list[i]+" error!")
            
        sentence_seged = jieba.cut(deal_raw.strip())    
        outstr = ''
        for word in sentence_seged:
            if word not in stopwords:
                if word != '\t' and word!='' and word!=' ':   #
                    outstr += word
                    outstr += " "
        #print(outstr)
        #print(type(outstr))
                
        f_in.close()
    
    elif factors_list[i][2]=='head':
        
        filepath=path_list[i]
        f_in = open(filepath,'r',encoding='utf-8') #osopen
        raw=csv.reader(f_in)
        head_c=0
        
        for last in raw:
            head_c=head_c+1
            if head_c==2:
                deal_raw=re.split("[{','}]", last[4])

        outstr=''
        for word in deal_raw:
            if word not in stopwords:
                if word != '\t' and word!='' and word!=' ':
                    outstr += word
                    outstr += " "
        
        f_in.close()
    
    
    outpath=path.join(outdir,name_list[i]+'.txt')
    
    #with open 
    f_out=open(outpath,'w',encoding='utf-8')
    f_out.write(outstr)
        
    f_out.close()           
    
    print(name_list[i]+" succeed!")
```
