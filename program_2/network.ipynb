```python
#读取权重数据
import pandas as pd
a=pd.read_csv("weight.csv")

df_w = pd.read_csv("weight.csv")

print(df_w.head())
```

```python
#将权重数据导入到图中
import networkx as nx
from networkx.algorithms import community as cmy
from networkx.algorithms.community import k_clique_communities
import matplotlib.pyplot as plt

data_w=df_w.values.tolist()
#print(data_w)

for i in data_w:
    G.node[i[0]]['weight']=i[1]
    G.node[i[0]]['effect']=0
    
weight= nx.get_node_attributes(G, 'weight')  
print(weight) 

effect= nx.get_node_attributes(G, 'effect')  
print(effect)       
```

```python
#检测双向边
for i in G:
    #print(i)
    for j in G:
        #若两个节点之间已有边
        if G.has_edge(i, j) and G.has_edge(j, i):
            print(str(i)+'+'+str(j))
            
print('over')
            
```

```python
#计算影响得分

process_count=0

for i in G:
    count=0
    for j in G:
        tmp_count=0
        tmp_d=0
        try:
            tmp1_distance=nx.dijkstra_path_length(G, source=i, target=j)
            tmp_count=1
        except:
            tmp1_distance=9999999
        try:
            tmp2_distance=nx.dijkstra_path_length(G, source=j, target=i)
            tmp_count=1
        except:
            tmp2_distance=9999999
            
        tmp_d=min(tmp1_distance,tmp2_distance)
        if tmp_d!=0 :
            G.node[i]['effect']=G.node[i]['effect']+G.node[j]['weight']/tmp_d
            count=count+tmp_count
    
    if count==0:
        count=1
        print(str(i)+'error')
    G.node[i]['effect']=G.node[i]['effect']/count
    process_count=process_count+1
    print(process_count)

effect= nx.get_node_attributes(G, 'effect')  
print(effect)       
```

```python
#计算最终得分
x=0.5  #占比调整参数

for i in G:
    G.node[i]['f_w']=G.node[i]['effect']*x+G.node[j]['weight']

f_w= nx.get_node_attributes(G, 'f_w')  
print(f_w) 
```
